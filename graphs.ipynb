```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.xlabel('time (seconds)')
plt.ylabel('User')
plt.yticks([0, 1, 2])

plt.axis([0, 2, 0, 2])
plt.plot([1], [1], 'ro')

plt.savefig('atomic.png')
plt.show()
```

```python
plt.xlabel('time (seconds)')
plt.ylabel('User')
plt.yticks([0, 1, 2])

x = np.arange(1., 5., 0.1)
y = np.ones(x.shape)
plt.plot(x, y, 'ro')
plt.axis([0, 6, 0, 2])

plt.savefig('timed.png')
plt.show()
```

```python
plt.xlabel('time (seconds)')
plt.ylabel('User')
plt.yticks([0, 1, 2, 3])

x = np.arange(1., 5., 0.1)

y1 = np.ones(x.shape)
plt.plot(x, y1, 'ro')

y2 = np.dot(np.ones(x.shape), 2)
plt.plot(x, y2, 'ro')

y3 = np.dot(np.ones(x.shape), 3)
plt.plot(x, y3, 'ro')

plt.axis([0, 6, 0, 4])

plt.savefig('atOnce.png')
plt.show()
```

```python
plt.xlabel('time (seconds)')
plt.ylabel('# concurrent users')
plt.yticks([0, 1, 2, 3])

x = [1, 2, 3, 4, 5]
y = [3, 3, 3, 3, 3]
plt.plot(x, y, 'k', x, y, 'yo')

plt.axis([0, 6, 0, 4])

plt.savefig('atOnce_users.png')
plt.show()
```

```python
plt.xlabel('time (seconds)')
plt.ylabel('User')
plt.yticks([0, 1, 2, 3])

x = np.arange(1., 5., 0.1)

y1 = np.ones(x.shape)
plt.plot(x, y1, 'ro')

x2 = x + np.dot(np.ones(x.shape), 1)
y2 = np.dot(np.ones(x.shape), 2)
plt.plot(x2, y2, 'ro')

x3 = x + np.dot(np.ones(x.shape), 2)
y3 = np.dot(np.ones(x.shape), 3)
plt.plot(x3, y3, 'ro')

plt.axis([0, 8, 0, 4])

plt.savefig('constantRate.png')
plt.show()
```

```python
plt.xlabel('time (seconds)')
plt.ylabel('# concurrent users')
plt.yticks([0, 1, 2, 3])

x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 2, 3, 3, 3, 2, 1]
plt.plot(x, y, 'k', x, y, 'yo')

plt.axis([0, 8, 0, 4])

plt.savefig('constantRate_users.png')
plt.show()
```

```python
plt.xlabel('time (seconds)')
plt.ylabel('User')
plt.yticks([0, 1, 2, 3, 4, 5, 6])

x = np.arange(1., 5., 0.1)

y1 = np.ones(x.shape)
plt.plot(x, y1, 'ro')

x2 = x + np.dot(np.ones(x.shape), 1)
plt.plot(x2, np.dot(np.ones(x.shape), 2), 'ro',
         x2, np.dot(np.ones(x.shape), 3), 'ro')

x2 = x + np.dot(np.ones(x.shape), 2)
plt.plot(x2, np.dot(np.ones(x.shape), 4), 'ro',
         x2, np.dot(np.ones(x.shape), 5), 'ro',
         x2, np.dot(np.ones(x.shape), 6), 'ro')

plt.axis([0, 8, 0, 7])

plt.savefig('rampRateInj.png')
plt.show()
```

```python
plt.xlabel('time (seconds)')
plt.ylabel('# concurrent users')
plt.yticks([0, 1, 2, 3, 4, 5, 6])

x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 3, 6, 6, 6, 5, 3]
plt.plot(x, y, 'k', x, y, 'yo')

plt.axis([0, 8, 0, 7])

plt.savefig('rampRateInj_users.png')
plt.show()
```

```python
plt.xlabel('time (seconds)')
plt.ylabel('User')
plt.yticks([0, 1, 2, 3, 4])

x = [1, 2, 3, 4, 5]
y = [1, 1, 1, 1, 1]
plt.plot(x, y, 'ro')
plt.plot(x, np.dot(y, 2), 'ro')
plt.plot(x, np.dot(y, 3), 'ro')

plt.axis([0, 6, 0, 4])

plt.savefig('load.png')
plt.show()
```

```python
plt.xlabel('time (seconds)')
plt.ylabel('# concurrent users')
plt.yticks([0, 1, 2, 3])

x = [1, 2, 3, 4, 5]
y = [3, 3, 3, 3, 3]
plt.plot(x, y, 'k', x, y, 'yo')

plt.axis([0, 6, 0, 4])

plt.savefig('load_users.png')
plt.show()
```

```python
plt.xlabel('time (seconds)')
plt.ylabel('User')
plt.yticks([0, 1, 2, 3, 4])

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [1, 1, 1, 1, 1, 1, 1, 1]
plt.plot(x, y, 'ro')
plt.plot(x[1:], np.dot(y, 2)[1:], 'ro')
plt.plot(x[2:], np.dot(y, 3)[2:], 'ro')

plt.axis([0, 9, 0, 4])

plt.savefig('load_ramp.png')
plt.show()
```

```python
plt.xlabel('time (seconds)')
plt.ylabel('# concurrent users')
plt.yticks([0, 1, 2, 3])

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [1, 2, 3, 3, 3, 3, 3, 3]
plt.plot(x, y, 'k', x, y, 'yo')

plt.axis([0, 9, 0, 4])

plt.savefig('load_ramp_users.png')
plt.show()
```
